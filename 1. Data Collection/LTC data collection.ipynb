```python
import requests, csv, time

symbol = 'LTCUSDT'
startTime = str(0)


for x in range(0,15):
            
    response = requests.get("https://api.binance.com/api/v3/klines?symbol=" + symbol + "&interval=1d&startTime=" +
                                startTime + "&limit=100")

    data = response.json()



    with open("ltc_data.csv", "a", newline='') as file:
        writer = csv.writer(file, delimiter=',')
        for d in data:
            writer.writerow(d)

    startTime = str(data[-1][6])

    time.sleep(1)
```

```python
import pandas as pd

df = pd.read_csv('ltc_data.csv', header=None)
df.info()
```

```python
df.columns = ['Open Time', 'Open', 'High', 'Low', 'Close', 'Volume ($)', 'Close Time', 'Quote Asset Volume (LTC)', 
              'No. of trades', 'Buy Base Volume (LTC)', 'Buy Quote Volume (LTC)', 'Ignore']
```

```python
df['Open Date'] = pd.to_datetime(df['Open Time'], unit='ms')
df['Close Date'] = pd.to_datetime(df['Close Time'], unit='ms')
```

```python
df['Close Date'] = df['Close Date'].dt.normalize()
```

```python
df.drop(['Open Time', 'Close Time', 'Ignore', 'Volume ($)', 'No. of trades', 'Buy Base Volume (LTC)', 'Buy Quote Volume (LTC)', 'Open Date'], axis=1, inplace=True)
```
