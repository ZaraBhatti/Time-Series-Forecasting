## I first use the following code to collect the data into a .csv file:

```python
import requests, csv, time

symbol = 'ETHUSDT'
startTime = str(0)


for x in range(0,15):
    
    
    
    response = requests.get("https://api.binance.com/api/v3/klines?symbol=" + symbol + "&interval=1d&startTime=" +
                                startTime + "&limit=100")

    data = response.json()



    with open("dataset.csv", "a", newline='') as file:
        writer = csv.writer(file, delimiter=',')
        for d in data:
            writer.writerow(d)

    startTime = str(data[-1][6])

    time.sleep(1)
        

```
